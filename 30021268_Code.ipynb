```python
def fetch_eikon_news (ric, start_date, end_date, max_items = 50):
    news = eikon.get_news_headlines(
        query= ric, 
        date_from= start_date, 
        date_to = end_date, 
        count = max_items
    )
   articles=[]
    for idx, row in news.iterrows():
       try:
        story = eikon.get_news_story(row["storyId"])
        articles.append({
            "ric":ric,
            "date":row["versionCreated"],
            "headline": row["text"],
            "story" : story
        })
    except:
        continue
    return pd.DataFrame(articles)
```
